4. Quais cidades têm a maior média de avaliações? E a menor média?

  a. Existe alguma explicação para isso? Conjecture.


```python
df_merged = pd.merge(df_details, df_ratings, on='hotel_id')

# Somando o número de avaliações por cidade
city_rating_count = df_merged.groupby('city_name')['number_of_ratings'].sum()

# Calculando o número total de imóveis por cidade
city_listing_total = df_details['city_name'].value_counts()

# Calculando a média de avaliações por cidade
city_average_ratings = city_rating_count / city_listing_total
city_average_ratings_sorted = city_average_ratings.sort_values()


cities_avg_ratings = city_average_ratings_sorted.index
avg_ratings = city_average_ratings_sorted.values

# Criando o gráfico
plt.figure(figsize=(10, 8))
plt.barh(cities_avg_ratings, avg_ratings, color='lightblue')
plt.xlabel('Média de Avaliações')
plt.ylabel('Cidades')
plt.title('Média de Avaliações por Cidade')
plt.grid(axis='x')

plt.show()
```

Considerando que "a maior média de avalições" se refere a quantidade de avaliações, é possível verificar no gráfico, a cidade com maior média de avaliações é Florianópolis com aproximadamente 113,74 avaliações por imóvel e a menor seria Imbituva com média de 0 avaliações.

a) Florianópolis ter média alta: Pode ser atribuída à popularidade da cidade como destino turístico, resultando em um maior número de visitantes e, consequentemente, um maior número de avaliações. Além disso, a qualidade dos imóveis e a satisfação dos hóspedes podem contribuir para a frequência de avaliações.

Imbituba e outras cidades com 0: A média de avaliações nula pode ser devido a vários fatores, como menor número de visitantes, novos imóveis ainda sem avaliações, ou talvez os imóveis em Imbituba não sejam tão populares ou bem classificados, levando a um menor número de avaliações.

6. Existem relações entre a nota do anúncio e os recursos disponíveis no imóvel?

  a. Quais? Explique.

```python
combined_df = pd.merge(df_ratings, df_details, on='hotel_id')

# Processando as comodidades
all_facilities_set = set()
for facilities_list in combined_df['room_facilities']:
    if isinstance(facilities_list, str):
        facilities_list = facilities_list.replace("[", "").replace("]", "").replace("'", "").split(", ")
        all_facilities_set.update(facilities_list)

# Convertendo o conjunto em uma lista
all_facilities = list(all_facilities_set)
all_facilities = [f for f in all_facilities if f]

# Criando um DataFrame temporário para as novas colunas
new_columns = pd.DataFrame()
for facility in all_facilities:
    new_columns[facility] = combined_df['room_facilities'].apply(lambda x: 1 if facility in x else 0)

# Concatenando as novas colunas
combined_df = pd.concat([combined_df, new_columns], axis=1)

# Calculando a correlação entre as comodidades e a nota total
correlation_with_total = combined_df.corrwith(combined_df['Total'], numeric_only=True)


correlation_df = pd.DataFrame({'Correlation': correlation_with_total})
correlation_df = correlation_df.dropna().sort_values(by='Correlation', ascending=False)

# Plotando o gráfico de correlação
plt.figure(figsize=(12, 15))
sns.heatmap(correlation_df, annot=True, cmap='coolwarm', linewidths=.5, cbar_kws={'label': 'Correlation'})
plt.title('Correlação das Comodidades com a Nota Total')
plt.show()
```

É possível concluir que existem relações entre as comodidades disponíveis em um imóvel e suas notas de avaliação. Algumas comodidades podem ser mais valorizadas pelos hóspedes, refletindo-se em notas mais altas, enquanto a ausência de certas comodidades pode afetar negativamente a avaliação. A análise também sugere que a quantidade total de comodidades pode ter um impacto na nota do imóvel.

Relação entre tamanho do imóvel e sua nota, existe alguma relação? Para isso dividi em Categorias: Grande, Médio e Pequeno.

```python
combined_df_7 = pd.merge(df_ratings, df_details, on='hotel_id', how='inner')

# Categorizando o tamanho do imóvel com base em porcentagem
small_size_limit = combined_df_7['room_surface_in_m2'].quantile(0.33)
large_size_limit = combined_df_7['room_surface_in_m2'].quantile(0.66)

def categorize_size(size):
    if size <= small_size_limit:
        return 'Pequeno'
    elif size <= large_size_limit:
        return 'Médio'
    else:
        return 'Grande'

combined_df_7['Size_Category'] = combined_df_7['room_surface_in_m2'].apply(categorize_size)

average_rating_by_size = combined_df_7.groupby('Size_Category')['Total'].mean()

plt.figure(figsize=(10, 6))
average_rating_by_size.plot(kind='bar')
plt.title('Nota Média por Categoria de Tamanho do Imóvel')
plt.xlabel('Categoria de Tamanho')
plt.ylabel('Nota Média')
plt.xticks(rotation=0)
plt.show()
```

Podemos ver que a média das notas quando relacionadas ao tamanho do imóvel não é um fator significativo.

Verificar se cidades com uma maior variedade de tipos de acomodação podem oferecer experiências mais diversificadas e talvez melhores avaliações?

```python
combined_df_7 = pd.merge(df_ratings, df_details, on='hotel_id', how='inner')

# Contando o número de tipos únicos de acomodação por cidade
accommodation_diversity = combined_df_7.groupby('city_name')['accommodation_type'].nunique()

# Calculando a média de notas por cidade
average_rating_by_city = combined_df_7.groupby('city_name')['Total'].mean()


diversity_and_ratings = pd.DataFrame({
    'Accommodation_Diversity': accommodation_diversity,
    'Average_Rating': average_rating_by_city
}).reset_index()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=diversity_and_ratings, x='Accommodation_Diversity', y='Average_Rating')
plt.title('Relação entre Diversidade de Acomodação e Notas Médias por Cidade')
plt.xlabel('Diversidade de Tipos de Acomodação')
plt.ylabel('Nota Média')
plt.show()
```

Baseado no gráfico de dispersão, podemos concluir que algumas cidades com menor diversidade de acomodações têm notas altas, enquanto outras com maior diversidade têm notas variadas. Sendo assim, cidades com maior diversidade de tipos de acomodação não mostram necessariamente notas médias mais altas. Isso sugere que a diversidade de tipos de acomodação em si pode não ser um fator determinante nas avaliações médias.
